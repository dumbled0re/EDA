```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = px.data.gapminder()
```

```python
file_path = '../dataset/countries_codes_and_coordinates.csv'
conuntry_master_df = pd.read_csv(file_path, sep=',', usecols=[2, 4, 5])
```

```python
# conuntry_master_dfのダブルクォーテーションを取り除く
for col in conuntry_master_df.columns:
    conuntry_master_df[col] = conuntry_master_df[col].apply(lambda x: x.strip()[1:-1])
```

```python
conuntry_master_df
```

```python
# conuntry_master_df.drop_duplicates()で重複の削除
pd.merge(df, conuntry_master_df.drop_duplicates(),
         left_on='iso_alpha', right_on='Alpha-3 code', how='inner')
```
