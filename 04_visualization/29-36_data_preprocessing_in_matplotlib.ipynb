```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
# figサイズを10,8に設定
plt.rcParams['figure.figsize'] = [10, 8]
plt.rcParams['font.size'] = 14
```

```python
file_path = '../dataset/house_price_jp.csv'
```

```python
df = pd.read_csv(file_path, sep=',')
```

```python
df.head()
```

```python
df['Log10_price'] = df['price'].apply(lambda x: np.log10(x))
```

```python
df.info()
```

### station_disとareaがobjectになっているのが気になる

```python
# station_disのユニークな値を確認
df['station_dis'].unique()
```

```python
# 数値に変換する(coerceで数値に変換出来ないものはnanに変換)
df['station_dis'] = pd.to_numeric(df['station_dis'], errors='coerce')
```

```python
df['station_dis'].unique()
```

```python
df['area'].unique()
```

```python
df['area'] = pd.to_numeric(df['area'], errors='coerce')
```

```python
df['area'].unique()
```

```python
# nanの削除
df = df.dropna(subset=['station_dis', 'area'])
```
