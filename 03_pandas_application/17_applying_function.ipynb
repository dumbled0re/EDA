```python
df['pop'].describe()
```

```python
# popの中央値
pop_med = 7.023596e+06
# 新しくpop_catという特徴量の作成(pop_medより大きい場合はgt_med、小さい場合はlt_medと代入する)
df['pop_cat'] = df['pop'].apply(lambda x: 'gt_med' if x >= pop_med else 'lt_med')
```

```python
df
```

```python
df['Log10_pop'] = df['pop'].apply(lambda x: np.log10(x))
```

```python
def add_continent_code(x):
    if x == "Asia":
        return 1
    elif x == "Europe":
        return 2
    else:
        return 3
```

```python
df['continent'].apply(add_continent_code)
```
