### 条件による抽出

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = px.data.gapminder()
```

```python
df
```

```python
# yearが1952のデータを取得
df["year"] == 1952
```

```python
df[df["year"] == 1952]
```

```python
gdp_per_cap_med = df["gdpPercap"].median()
condition1 = df["gdpPercap"] > gdp_per_cap_med
condition2 = df["year"] >= 2000
df[condition1 & condition2]
```

```python
# queryを使用したデータ抽出
df.query("year >= 2000")
```

```python
df.query("(year >= 2000) & (continent == 'Asia')")
```

```python
df["continent"].unique()
```

```python
df.query("continent in ['Asia', 'Europe', 'Africa']")
```

```python
target_continents = ['Asia', 'Europe', 'Africa']
# @で変数名を使用可能
df.query("continent in @target_continents")
```

```python
target_continents = ['Asia', 'Europe', 'Africa']
df["continent"].isin(target_continents)
```

```python
df[df["continent"].isin(target_continents)]
```

```python
# countryの中のAを含んでいればTrueを返す
df["country"].str.contains("A")
```

```python
df[df["country"].str.contains("A")]
```

### Null値の処理

```python
df
```

```python
df.loc[1704, "continent"] = "Asia"
df
```

```python
df.info()
```

```python
# null値を0で埋める
df.fillna(0)
```

```python
df.fillna(1)
```

```python
# 特定のカラムのnull値を特定の値に置き換える
df.fillna({"country": "No country", "year": 1952})
```

```python
# null値があるレコードを削除する
df.dropna()
```

```python
# レコードの値がすべてnull値の場合のみ削除(continentにAsiaが含まれるので何も削除されない)
df.dropna(how="all")
```

```python
# 特定のカラムにnull値がある場合はそのレコードを削除
df.dropna(subset=["year"])
```

```python
# continentにはnull値がないので、どのレコードも削除されない
df.dropna(subset=["continent"])
```

### 重複データの処理

```python
# 重複データ作成
df.loc[1704, :] = df.loc[1703, :]
df
```
