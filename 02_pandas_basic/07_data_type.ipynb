### データの生成・消去

```python
# カラム名とデータを生成する
df["flag"] = 1
```

```python
df.head()
```

```python
# 全カラムの最後の行に1を入れる
df.loc[1704, :] = 1
```

```python
df.tail(5)
```

```python
# データの削除(インデックスを指定)
df = df.drop(1704)
# inplaceは新しくdfに代入しなくても大丈夫になる(df = df.drop("flag", axis=1)と同じ効果)
df.drop("flag", axis=1, inplace=True)
```

```python
# データ型の変換
df["year"] = df["year"].astype(str)
```

```python
# 元に戻す
df["year"] = df["year"].astype(float)
```

```python
df.loc[1704, "year"] = "abc"
```

```python
df
```

```python
# yearカラムにはabcがあるので数字型に変換するとエラーが発生するがto_numeric関数を使用するとabcがNaNに変わる
df["year"] = pd.to_numeric(df["year"], errors="coerce")
```
